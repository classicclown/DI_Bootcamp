```python
from scipy import stats
import scipy as sp
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

data = [12, 15, 13, 12, 18, 20, 22, 21]

mean = sp.mean(data)
median = sp.median(data)
variance = sp.var(data)
std = sp.std(data)
```

```python
mean = 50
std = 10
dist = stats.norm(mean, std)
data = dist.rvs(1000)
plt.hist(data, bins=20)
plt.show()

```

```python
array1 = np.array([1, 2, 3, 4, 5])
array2 = np.array([6, 7, 8, 9, 10])

ttest , pval = stats.ttest_ind(array1, array2)
print(ttest, pval)
```

```python
alpha = 0.05
if pval < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

```python
data1 = np.random.normal(0, 1, 1000)
data2 = np.random.normal(0, 1, 1000)

corr = stats.pearsonr(data1, data2)
print(corr)

skewness = stats.skew(data1)
print(skewness)

#Two datasets are negatively correlated but the P-value indicates that this is highly due to randomness
#The skewness indicates that the data is very slightly positively skewed, meaning the tail extends to the positive values.
```
