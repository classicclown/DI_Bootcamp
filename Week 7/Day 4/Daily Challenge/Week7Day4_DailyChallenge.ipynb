```python
import matplotlib.pyplot as plt
from PIL import Image, ImageOps
```

```python
img = Image.open('/content/flower_images/flower_images/0001.png')
```

```python
def show_images(original, augmented, title):
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.imshow(original)
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(augmented)
    plt.title(title)
    plt.axis('off')

    plt.show()
```

```python
rotated_img = img.rotate(90)
show_images(img, rotated_img, 'Rotated Image')
```

```python
horizontally_flipped_img = ImageOps.mirror(img)
vertically_flipped_img = ImageOps.flip(img)
show_images(img, horizontally_flipped_img, 'Horizontally Flipped Image')
show_images(img, vertically_flipped_img, 'Vertically Flipped Image')
```

```python
width, height = img.size
new_width = int(width * 1.2)
new_height = int(height * 1.2)
resized_img = img.resize((new_width, new_height))
show_images(img, resized_img, 'Resized Image 1.2x')
```
