Analysing raw numbers is not very useful to us, we cannot understand data or identify patterns in the numbers just by looking at them. We are visual creatures by nature so using visualisations is neccesary to digest and understand data at a glance.


A line graph is useful to show a change in data over time, like change in temps from 2022 to 2023.

```python
import matplotlib.pyplot as plt

temp= [72, 74, 76, 80, 82, 78, 75]
days = [1, 2, 3, 4, 5, 6, 7]

plt.plot(days, temp)
plt.xlabel('Days')
plt.ylabel('Temp (F)')
```

```python
sales=[5000, 5500, 6200, 7000, 7500]
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']

plt.plot(months, sales)
plt.xlabel('Months')
plt.ylabel('Sales')
```

```python
import pandas as pd
url = 'https://raw.githubusercontent.com/intern2grow/sales-data-analysis/main/sales_data.csv'
df = pd.read_csv(url)
df.head()
```

```python
total_products = df['quantity'].sum()
total_products
```

```python

grouped_df = df.groupby('category').agg({'revenue': 'sum'})
grouped_df.sort_values('revenue', ascending=False)
```

```python
avg_rev = df['revenue'].mean()
avg_rev
```

```python
df['date']= pd.to_datetime(df['date'])
quarterly_data = df.resample('Q', on='date').sum()
df['category'].replace({'Bgas:Bags','Shoeses:Shoes','Clohting:Clothing'})
quarterly_data.index = quarterly_data.index.to_period('Q').strftime('Q%q')

quarterly_data
```

```python
plt.bar(quarterly_data.index, quarterly_data['revenue'])
plt.xlabel('Quarter')
plt.ylabel('Revenue')
```

```python
import numpy as np

x=np.linspace(-10,10,20)
y=(x**2)

plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('My function')
```

```python
sales_data = [15,30,45,20]
categories=['A','B', 'C', 'D']

plt.bar(categories, sales_data)
plt.xlabel('Categories')
plt.ylabel('Sales')
plt.title('Sales by Category')
```

```python
fruit = ['Apples','Bananas','Cherries','Dates']
prefs = [0.1,0.3,0.2,0.15]
plt.pie(prefs, labels=fruit)
plt.title('My Preferences')
plt.legend(loc='upper right')
```
