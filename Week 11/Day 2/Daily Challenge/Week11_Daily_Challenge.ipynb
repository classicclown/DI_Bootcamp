```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
temp_data = np.random.uniform (-5,35,(10,12))
cities =['Cape Town','Tel Aviv','Montreal','New York','Sydney','Oaxaca','London','Amsterdam','Brussels','Kyiv']
months=['January','February','March','April','May','June','July','August','September','October','November','December']
temp_df = pd.DataFrame(temp_data,index=cities, columns=months)
temp_df.head()
```

```python
mean_temp=temp_df.mean(axis=1).round(2)
annual_mean_temp=mean_temp.mean().round(2)
highest_temp_city = mean_temp.idxmax()
lowest_temp_city = mean_temp.idxmin()

print(f'Mean Temp:\n{mean_temp}')
print(f'Annual Mean Temp: {annual_mean_temp}')
print(f'Highest Temp City: {highest_temp_city}')
print(f'Lowest Temp City: {lowest_temp_city}')
```

```python
#sns.set(rc = {'figure.figsize':(15,7)})
plt.figure(figsize=(15,7))
line_plot=sns.lineplot(temp_df)
line_plot.set_title('Monthly Temperature Trends for Each City')
line_plot.set_xlabel('Month')
line_plot.set_ylabel('Temperature')
```
