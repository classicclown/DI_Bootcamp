**Objective: Create a line plot using Plotnine that compares two different data series.**

Create a pandas DataFrame with three columns: x, y1, and y2, where x is a sequence of numbers and y1, y2 are two different mathematical functions of x (e.g., sin(x) and cos(x)).

-Plot y1 and y2 against x in the same plot, using different colors for each line.
-Add a legend to distinguish between the two lines, and label axes as ‘X-Value’, ‘Y-Value’, and add a title ‘Comparison of Two Functions’.

-Expected Outcome:

A line plot that effectively compares the two functions across the range of x, with a legend, axis labels, and a title.


```python
x = np.linspace(0, 10, 50)

y1 = np.sin(x)
y2 = np.cos(x)

df = pd.DataFrame({'x': x, 'y1': y1, 'y2': y2})

df_melt = df.melt(id_vars=['x'], value_vars=['y1','y2'], var_name='color', value_name='y1_and_y2')


df_melt.head()
```

```python
line_plot = ggplot(df_melt)+geom_line(aes(x='x',y='y1_and_y2',color='color'))+labs(x='X',y='Y-series',title='Series 1 vs Series 2')
line_plot
```

Create a basic line chart using Plotly to visualize the trend of sales over 12 months. Assume the following monthly sales data:




months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales = [200, 210, 250, 300, 320, 340, 360, 380, 390, 400, 420, 440]


```python
import plotly.express as px

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales = [200, 210, 250, 300, 320, 340, 360, 380, 390, 400, 420, 440]

df = pd.DataFrame({'Months':months, 'Sales':sales})
df.head()
```

```python
line_chart = px.line(df, x='Months', y='Sales')
line_chart.show()
```

Using Plotly, create a bar chart to represent the number of books sold in different genres. Use the following data:




genres = ['Fiction', 'Non-Fiction', 'Sci-Fi', 'Mystery', 'Biography']
books_sold = [120, 150, 90, 80, 130]


Tasks:

Import Plotly’s graph_objects as go.
Create a bar chart with genres on the x-axis and books_sold on the y-axis.
Customize the bar colors.
Add a title to the chart: “Books Sold by Genre”.
Display the plot.


```python
import plotly.graph_objects as go


genres = ['Fiction', 'Non-Fiction', 'Sci-Fi', 'Mystery', 'Biography']
books_sold = [120, 150, 90, 80, 130]
df = pd.DataFrame({'genres':genres, 'books_sold':books_sold})
colors = ['red', 'green', 'blue', 'orange', 'purple']


fig = go.Figure()
fig.add_trace(go.Bar(x=df['genres'],y=df['books_sold'],marker=(dict(color=colors))))
fig.update_xaxes(title_text="Genres")
fig.update_yaxes(title_text="Number Sold")
```

Create a scatter plot to visualize the relationship between Age, Salary, and Years of Experience in a dataset. Assume the following data:




import pandas as pd

data = pd.DataFrame({
    'Age': [25, 30, 35, 40, 45],
    'Salary': [50000, 60000, 70000, 80000, 90000],
    'Years of Experience': [2, 5, 7, 10, 12]
})


Tasks:

Import Plotly’s express as px.
Create a scatter plot with Age on the x-axis, Salary on the y-axis.
Use Years of Experience to determine the size of each marker.
Add a title to the chart: “Relationship between Age, Salary, and Experience”.
Display the plot.


```python
import plotly.express as px

data = pd.DataFrame({
    'Age': [25, 30, 35, 40, 45],
    'Salary': [50000, 60000, 70000, 80000, 90000],
    'Years of Experience': [2, 5, 7, 10, 12]
})

scatter_plot = px.scatter(data, x='Age',y='Salary',size='Years of Experience', title='Relationship between Age, Salary, and Experience')
scatter_plot.show()
```
